```python
import module
```

```python
TrainData = module.pd.read_csv(r"C:\Users\adity\OneDrive\Desktop\NNFS\Data\mnist_train.csv")
df = module.pd.DataFrame(TrainData)
df
```

```python
X = df.drop("label", axis=1).values
Y = df["label"].values   
X = X / 255.0
```

```python
Y_encoded = module.one_hot(Y)
Y
```

```python
W1, b1, W2, b2, W3, b3 = module.init_params()
weights = [W1, W2, W3]
biases = [b1, b2, b3]
```

```python
epochs = 30
batch_size = 32
learning_rate = 0.1
num_batches = X.shape[0] // batch_size 

for epoch in range(epochs):
    epoch_loss = 0

    for i in range(0, X.shape[0], batch_size):
        X_batch = X[i:i+batch_size]
        Y_encoded_batch = Y_encoded[i:i+batch_size]

        weights, biases, batch_loss = module.forward_backward_pass(X_batch, Y_encoded_batch, weights, biases)
        epoch_loss += batch_loss
    print(f"Epoch {epoch+1}/{epochs}, Loss: {epoch_loss/num_batches}")
```

```python
Test_Data = module.pd.read_csv(r"C:\Users\adity\OneDrive\Desktop\NNFS\Data\mnist_test.csv")
tdf = module.pd.DataFrame(Test_Data)
X_test = tdf.drop("label", axis=1).values
Y_test = tdf["label"].values   
X_test = X_test / 255.0

```

```python
Y_pred_test = module.predict(X_test, weights, biases)
accuracy = module.np.mean(Y_pred_test == Y_test)
print("Training Accuracy:", accuracy)
```
